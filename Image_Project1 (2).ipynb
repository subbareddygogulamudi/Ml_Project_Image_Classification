```python
df=pd.read_csv("/content/Blood_Cell_Images.csv")
```

Let us assign all features to the x variable and target to the y variable.

```python
df.info()
```

```python
x=df.drop("Target",axis=1)
y=df["Target"]
```

Checking the shape of x and y

```python
print(x.shape)
print(y.shape)
```

Splitting the data into train and test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Building a model with KNN algorithm

```python
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
knn_pipe = Pipeline(steps = [("knn",KNeighborsClassifier(n_neighbors=3))])
knn_pipe.fit(x_train,y_train)
```

```python
y_pred = knn_pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
score = accuracy_score(y_pred,y_test)
print(score)
```

Building a model with Naive Bayes Algorithm

```python
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import GaussianNB
nb_pipe = Pipeline(steps = [("NB",GaussianNB())])
nb_pipe.fit(x_train,y_train)
```

```python
y_pred =nb_pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
score = accuracy_score(y_pred,y_test)
print(score)
```

Building a model with Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
dt_pipe=Pipeline(steps = [("dt",DecisionTreeClassifier())])
dt_pipe.fit(x_train,y_train)
y_pred =dt_pipe.predict(x_test)
score = accuracy_score(y_pred,y_test)
print(score)
```

Hyper Parameter tuning using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
tuned_parameters=[{"max_depth":[i for i in range(1,21)]}]
dt=GridSearchCV(
    estimator=DecisionTreeClassifier(),
    param_grid=tuned_parameters,
    scoring="accuracy",
    cv=5,
    return_train_score=True,
    verbose=1
)
```

```python
dt.fit(x_train,y_train)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr_pipe=Pipeline(steps=[("lr",LogisticRegression(max_iter=100))])
lr_pipe.fit(x_train,y_train)
y_pred=lr_pipe.predict(x_test)
score=accuracy_score(y_pred,y_test)
print(score)
```

Support Vector Machine

```python
from sklearn.svm import SVC
svm_pipe=Pipeline(steps=[("svm",SVC())])
svm_pipe.fit(x_train,y_train)
```

```python
y_pred=svm_pipe.predict(x_test)
score=accuracy_score(y_pred,y_test)
print(score)
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf_pipe=Pipeline(steps=[("rf",RandomForestClassifier(
    n_estimators=300,
    max_depth=25,
    min_samples_split=5,
    min_samples_leaf=2,
    n_jobs=-1,
    random_state=42
))])
rf_pipe.fit(x_train,y_train)
```

```python
y_pred=rf_pipe.predict(x_test)
score=accuracy_score(y_pred,y_test)
print(score)
```

Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
gb_pipe=Pipeline(steps=[("gb",GradientBoostingClassifier())])
gb_pipe.fit(x_train,y_train)
```

```python
y_pred=gb_pipe.predict(x_test)
score=accuracy_score(y_pred,y_test)
print(score)
```
